## Check relationship with target

```python
# check vars IV
def compute_IV(data_raw, predictors, target_col):
    
    output={}
    
    for var in predictors:
            data=data_raw[[target_col, var]].copy()
            data.columns=['label', var]
            base = data.assign(good=np.where(data['label']==0, 1, 0))\
                    .assign(bad=np.where(data['label']==1,1,0))

            weight = 1
        
            base['good']=base['good'] * weight
            base['bad']=base['bad'] * weight
            base['weight']=weight
            total_good = sum(base['good'])
            total_bad = sum(base['bad'])

            temp_data = base.groupby(var).agg(
                    good = pd.NamedAgg(column='good', aggfunc='sum'),
                    bad = pd.NamedAgg(column='bad', aggfunc='sum'),
                    total = pd.NamedAgg(column='weight',aggfunc='sum')
                    )
            temp_data['good_pct']=temp_data['good']/total_good
            temp_data['bad_pct']=temp_data['bad']/total_bad
            temp_data = temp_data.assign(woe=lambda x: np.log(x.good_pct/x.bad_pct))
            temp_data = temp_data.assign(IV =lambda x: x.woe*(x.good_pct - x.bad_pct))

            # temp_data['woe_rank'] = temp_data['woe'].rank(method='min')

            output[var] = temp_data
        
            del temp_data, data, base

    return output
```

```python
df_IV = df.copy()
```

```python
varlist=['revol_util_n', 'installment_n', 'int_rate_n', 'emp_length_n', 'dti_n', 'annual_inc_n']

for v in varlist:
    df_IV[v] = pd.qcut(df_IV[v], 5)    
    result_IV=compute_IV(df_IV, [v], "Outcome")
    print(result_IV)
    
```
